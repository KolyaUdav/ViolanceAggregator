```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import json
```

```python
json_files = 0

for _,_,files in os.walk('../json/'):
    json_files = files

data_arr = []
    
for jf in json_files:
    data_dict = {}
    
    with open('../json/{}'.format(jf), 'r') as data_json:
        data_dict = json.load(data_json)
    
    data_arr.append(data_dict)
    
print(data_arr)
```

```python
column_names = {'возраст': 'age',
                'пренебрежение_нуждами': 'Пренебрежение нуждами ребёнка',
                'физическое_насилие_дома': 'Физическое насилие дома',
                'физическое_насилие_в_школе': 'Физическое насилие в школе',
                'психологическое_насилие_дома': 'Психологическое насилие дома',
                'психологическое_насилие_в_школе': 'Психологическое насилие в школе',
                'сексуальное_насилие_дома': 'Сексуальное насилие дома',
                'сексуальное_насилие_в_школе': 'Сексуальное насилие в школе'}

df = pd.DataFrame(columns=[k for k in column_names.keys()])

for d in data_arr:
    for value in d.items():
        sv = value[1]['specific']
        cv = value[1]['categories']

        df = df.append({k:sv[v] if k == 'возраст' else len(cv[v]) for (k, v) in column_names.items()}, ignore_index=True)
```

```python
len(df[df['физическое_насилие_в_школе'] > 0])
```

```python
total_percents_dict = {}

df = df.drop(['возраст'], axis=1)

for c in df.columns.values.tolist():
    total_percents_dict[c] = round((len(df[df[c] > 0]) / len(df[c])) * 100)
```

```python
df_total_percents = pd.DataFrame(total_percents_dict, index=['total_percents_violanced'])
```

```python
for c in df_total_percents.columns.values.tolist():
    total_percents_dict[c] = round(((len(df[c]) - len(df[df[c] > 0])) / len(df[c]) * 100))
```

```python
df_total_percents
```

```python
df_total_percents = df_total_percents.append(pd.DataFrame(total_percents_dict, index=['total_percents_not_violanced']))
```

```python
pie_data = [
    df_total_percents['физическое_насилие_дома']['total_percents_violanced'],
    df_total_percents['физическое_насилие_дома']['total_percents_not_violanced']
]
    
pie_names = ['Имелись факты насилия', 'Эпизоды насилия отсутствуют']
    
colors = ['r', 'b']
    
plt.pie(pie_data, labels=pie_names, colors=colors)

plt.show()
```

```python
df_violance_age = df.drop(['пол', 'семья', 'родственники', 'средовое_окружение'], axis=1)
```

```python
age_array = ['10', '11', '12', '13', '14', '15', '16', '17']

age_total_percents_dict = {}
for c in df_violance_age.columns.values.tolist():
    if c != 'возраст':
        age_total_percents_dict[c] = {}
        for age in age_array:
            age_total_percents_dict[c][age] = round((len(df_violance_age[(df_violance_age[c] > 0) & (df_violance_age['возраст'] == age)]) / len(df_violance_age[df_violance_age['возраст'] == age])) * 100)
        
        
```

```python
df_age_total_percents = pd.DataFrame(age_total_percents_dict)
```
